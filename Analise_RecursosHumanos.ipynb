```python
#Libs Essenciais

# Libs para modelagem e matrizes
import numpy as np
import pandas as pd

# Libs para análises
import matplotlib.pyplot as plt
import seaborn as sns

# Libs para ignorar avisos
import warnings

# Desabilitando os avisos
warnings.filterwarnings('ignore')


#Lendo o dataframe
Base_Dados = pd.read_csv('Salary_Data.csv')

#Verificando
Base_Dados.head()
```

```python
# Renomeando as colunas
Base_Dados.rename( columns={
    'YearsExperience': 'Anos de Experiencia',
    'Salary': 'Salario'
}, inplace=True)
```

```python
#Dimensao
Base_Dados.shape
```

```python
#Campos nulos
Base_Dados.isnull().sum()
```

```python
#Gráfico de Nulos
sns.heatmap(Base_Dados.isnull(), cbar=False)

#Obs: Podemos concluir que as linhas e colunas estão todas preenchidas
```

```python
Base_Dados.describe()

#Obs: podemos notar que temos desde 1 a 10 anos de experiencia na empresa, e a média salarial é 37.000 U$D/Ano.
```

```python
#Calculando média mensal

37000/12

#Obs:A média salarial é de 3.083 dólares por mês nesta empresa
```

```python
#Gráfico de Renda

plt.figure( figsize=(10,8))
plt.title('Análise de Renda dos Funcionários')
sns.kdeplot( Base_Dados['Salario'], color='red', linewidth=2, shade=True)
plt.show()
```

```python
#Gráfico de Experiência

plt.figure( figsize=(10,8))
plt.title('Análise de Experiência dos Funcionários')
sns.kdeplot( Base_Dados['Anos de Experiencia'], color='blue', linewidth=2, shade=True)
plt.show()
```

```python
#Análise de Distribuição - Salário

plt.figure( figsize=(10,4))
sns.boxplot( Base_Dados['Salario'])
```

```python
#Análise de Distribuição - Experiência

plt.figure( figsize=(10,4))
sns.scatterplot( data=Base_Dados,x='Anos de Experiencia', y='Salario')

#Obs: Quanto mais tempo de experiencia, maior o salário do funcionáriuo, é o que podemos observar com este gráfico comparativo
```

```python
sns.regplot( data=Base_Dados,x='Anos de Experiencia', y='Salario')

#Obs: É nítido a relação entre as duas variaveis
```

```python
#Analise de Correlação
Correlacao = Base_Dados.corr('pearson')
plt.figure( figsize=(5,5))
sns.heatmap( Correlacao, annot=True, cbar=False);
```

```python
#Obs: A correlação de Pearson é mais recomendada para analisar dados em padrão de distribuição linear
```
